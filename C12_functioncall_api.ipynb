```python
import tensorflow as tf
from keras import Input, layers, models

# 单一输入和输出

# 输入层
inputs = Input(shape=(4,))
# 全连接层 堆叠
x = layers.Dense(64,activation='relu')(inputs)
# 输出层
outputs = layers.Dense(10,activation='softmax')(x)

model = models.Model(inputs=inputs, outputs=outputs)

# 查看模型结构
model.summary()
```

```python
# 多个输入和输出
# 输入层1
inputA = Input(shape=(64,), name='inputA')
# 输入层2
inputB = Input(shape=(128,), name='inputB')
# 全连接层1
x = layers.Dense(32, activation='relu')(inputA)
# 全连接层2
y = layers.Dense(32, activation='relu')(inputB)
# 合并层
combined = layers.concatenate([x, y])
# 输出层1
z1 = layers.Dense(1, activation='sigmoid', name='output1')(combined) # 二分类
# 输出层2
z2 = layers.Dense(10, activation='softmax', name='output2')(combined) # 多分类
# 构建模型
model = models.Model(inputs=[inputA, inputB], outputs=[z1, z2])
# 查看模型结构
model.summary() # 支持了多输入多输出
```
