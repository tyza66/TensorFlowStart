```python
# 鸢尾花分类问题
import tensorflow as tf
from keras import Input, layers
from sklearn.datasets import load_iris
# 1，加载鸢尾花数据集
iris = load_iris()
X = iris.data # 特征：花萼长度、花萼宽度、花瓣长度、花瓣宽度
y = iris.target # 标签：0-Setosa, 1-Versicolour, 2-Virginica
```

```python
# 构建模型
model = tf.keras.models.Sequential([
Input(shape=(X.shape[1],)), # 输入层
layers.Dense(16, activation='relu'), # 隐藏层 能加速
layers.Dense(3, activation='softmax') # 输出层 3个神经元，对应3个类别
])
model.summary()
```

```python
import datetime

# 定义 TensorBoard 日志目录
log_dir = "logs/tf2fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
# 编译模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# 多分类交叉熵损失函数 适合整数标签这种情况 如果想要 onehot 的标签 可以用 categorical_crossentropy
```

```python
# 模型训练
history = model.fit(X, y, epochs=100, batch_size=32, verbose=1,callbacks=[tensorboard_callback])
print(f"最终损失: {history.history['loss'][-1]:.4f}, 最终准确率:{history.history['accuracy'][-1]:.4f}")
```
